# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

The first graph, with two separate panels, is better for comparing each week’s pattern independently. It is cleaner, avoids overlap, and makes the shape of each time series easier to read. The downside is that direct comparison between the two weeks is a little less immediate, because your eyes have to move between panels.

The second graph, with both lines on the same axes, makes comparison between the two weeks much easier. You can immediately see which week is higher on each day and where the trends cross. The downside is that it can be harder to read when the lines overlap or get close together, and it becomes less clear if there are many series or more complex data.
